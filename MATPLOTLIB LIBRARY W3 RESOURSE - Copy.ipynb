```python
import numpy as np
import matplotlib.pyplot as plt
x = [10,20,30]
y = [20,40,10]
plt.plot(x,y, 'b', label = 'line 1')
x2 = [10,20,30]
z = [40,10,30]
plt.plot(x2,z,'g', label = 'line 2')
plt.legend()
```

```python
x = [10,20,30]
y = [20,40,10]
z = [40,10,30]
plt.plot(x,y,'b',linewidth = 3, label = 'line1 - width - 3')
plt.plot(x,z,'r', linewidth = 5, label = 'line2_ width -5')

plt.legend()
```

```python
x = [10,20,30]
y = [20,40,10]
z = [40,10,30]
plt.plot(x,y,'b',linewidth = 3, linestyle = 'dotted',label = 'line1 - dotted')
plt.plot(x,z,'r', linewidth = 2,linestyle = 'dashed',label = 'line2_ dashed')

plt.legend()
```

# BAR PLOT

```python
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
x_pos = [i for i, _ in enumerate(x)]
plt.bar(x_pos, popularity, color='blue')
```

```python
import matplotlib.pyplot as plt
a = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
b = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
plt.xlim(0,25)
plt.barh(a,b, color = 'g')
plt.minorticks_on()
plt.grid(which = 'major',color = 'r',linewidth = '0.5')
plt.grid(which = 'minor',linestyle = ':')
```

```python
a = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
b = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
plt.ylim(0,25)
plt.bar(a,b, color = ['r','black','g','b','y','c'])
plt.minorticks_on()
plt.grid(which = 'major',color = 'r',linewidth = '0.5')
plt.grid(which = 'minor',linestyle = ':')
```

# pie charts

```python
import matplotlib.pyplot as plt
languages = 'Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'
popuratity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
c = ['r','g','b','black','yellow','orange']
explode = (0.01, 0.01, 0.01, 0.01,0.01,0.01) 
plt.pie(popuratity, colors = c,labels =languages, explode = explode,startangle = 90, autopct = '%1.1f%%')
plt.show
```

# scatter plot

```python
import matplotlib.pyplot as plt
from pylab import randn
x = randn(200)
y = randn(200)
plt.scatter(x,y, color = 'r')
plt.show()
```

```python
import matplotlib.pyplot as plt
from pylab import randn
x = randn(200)
y = randn(200)
plt.scatter(x,y, edgecolors = 'r', s = 70, facecolors = 'g', color = 'b')
plt.show()
```

```python
sns.jointplot(x = 'total_bill',y = 'tip', data = tips, kind = 'hex')
```

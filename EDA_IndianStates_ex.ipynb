# Histogram plot of Sex Ratio.

```python
x = data['sex_ratio']
plt.hist(x)
plt.show()
```

```python
figure = plt.figure(figsize=(15,8))
plt.hist(data['sex_ratio'],color = 'olive',bins = 30)
plt.xlabel('sex_ratio')
plt.ylabel('count of cities')
plt.title('Histogram of Sex-ratio')
plt.show()
```

```python
x = data['child_sex_ratio']
plt.hist(x)
plt.show()
```
